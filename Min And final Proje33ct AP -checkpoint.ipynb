```python
import openpyxl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
#project


def GetExcelFile(File):
    
    #Get EXCAL FILE THORGH OPEN LIBRAY
    return openpyxl.load_workbook(File)
def GetGraphTeg(Clash_Report):
    GraphTeg = list()
    for Team in Clash_Report.iter_cols(0, NumTeams):

        # FOR TOP PART
        GraphTeg.append(Team[0].value)

    return GraphTeg
def GetMetchLocation(Clash_Report):

    TeamLocater = list()
    TeamAndLocaterMap = dict()
    Locater= 0

    for Team1 in Clash_Report.iter_cols(0, NumTeams):      
        for Match1 in range(1, NumMatchs):
            if Team1[Match1].value is None:
                pass           
            else:
                Locater = TeamAndLocaterMap.get(Team1, Locater)
                TeamAndLocaterMap[Team1] = Locater

                Information = Team1[Match1].value
                
                for Team2 in Clash_Report.iter_cols(0, NumTeams):
                    if Team2 != Team1:
                        for Match2 in range(1, NumMatchs):
                            if Team2[Match2].value is None:
                                pass
                            else:
                                if Information  == Team2[Match2].value:
                                    TeamLocater.append(Locater)
                    Locater  += 1

    return TeamLocater

def MetrixSize(NumMatchs, NumTeams):
    return NumTeams * NumTeams
def GetMetrixLocater(NumMatchs, NumTeams):
    MetrixLocater = list()
    Position = 0
    for Match in range(NumTeams):
        for Team in range(NumTeams):
            if Match == Team:
                MetrixLocater.append(Position)
            Position += 1

    return MetrixLocater
def GetDataMetrixGrid(CompletGrid, TeamLocater, NumMatchs, NumTeams):
    Data = np.zeros(CompletGrid)
    Data[TeamLocater] = np.ones(len(TeamLocater))

    MetrixLocater = GetMetrixLocater(NumMatchs, NumTeams)
    MetrixLocaterInfo = [2] * len(MetrixLocater)

    Data[MetrixLocater] = np.array([MetrixLocaterInfo])

    Data = np.ma.array(Data.reshape((NumTeams, NumTeams)), mask=Data == 0)

    return Data
def DesplayGraph(Data, GraphTeg):
    # BULID DICARDED COLORMAP
    cmap = colors.ListedColormap(['green', 'black'])
    fig, ax = plt.subplots(figsize=(20,20))
    ax.imshow(Data, cmap=cmap)
    fig.autofmt_xdate()
    # OPTIONALLY INCLUDE GEID 
    ax.set_xticks(np.arange(NumTeams) - 0.51, labels=GraphTeg)
    ax.set_yticks(np.arange(NumTeams) - 0.51, labels=GraphTeg)
    ax.grid(which="major", color='grey', linewidth=1)
    #ax.grid(which="major", color='grey', linewidth=1, axis='both', linestyle='-')
    
    #ax.grid(which='major', , )
    ax.tick_params(which="minor",)

    plt.show()


DataFrame = GetExcelFile("Clash_Report.xlsx")
Clash_Report = DataFrame.active

# ACHEVED COMPLET  NUMBER OF ROWS AND NUMBER COLUMNS 
NumMatchs = Clash_Report.max_row
NumTeams = Clash_Report.max_column
# Get MATCH LOCATION
TeamLocater = GetMetchLocation(Clash_Report)
# GET COMPLET METRIX
CompletGrid = MetrixSize(NumMatchs, NumTeams)
# GET OVER ALLGRAP TEG TEAMS AND MATCHS 
GraphTeg = GetGraphTeg(Clash_Report)

# MATRIX DATA TO DEDPLAY IN THE GRID
Data = GetDataMetrixGrid(CompletGrid, TeamLocater, NumMatchs, NumTeams)
# DESPLAY AND GRAPH TEG
DesplayGraph(Data, GraphTeg)
```
